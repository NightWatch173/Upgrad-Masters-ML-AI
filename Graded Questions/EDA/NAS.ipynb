```python
df1 = df[['Mother.edu', 'Siblings']].copy()
df1.head()
```

```python
df1['Siblings'].unique()
```

```python
def clean_siblings(x):
    if x in ['1 sibling', '3 siblings', '2 siblings', '4+ siblings']:
        return int(str(x)[0])
    else:
        return 0
```

```python
df1['Siblings'] = df1['Siblings'].apply(lambda x: clean_siblings(x))
df1.head()
```

```python
df1.groupby('Mother.edu').agg(lambda x: x.value_counts().index[0])
```
